```python
import pandas as pd
import numpy as np
import pyomo as pyo
import openpyxl
from pathlib import Path
```

```python
data_folder_path = "00-INPUT-DATA/1-complete_data-mixed-all-1-sw"

folder = Path(data_folder_path)
print(f"Contents of: {folder.resolve()}")

for p in sorted(folder.iterdir()):
        kind = "DIR " if p.is_dir() else "FILE"
        print(f"{kind}  {p.name}")

```

```python
csv_files = sorted(folder.glob("*.csv"))

for csv_file in csv_files:
    df_name = csv_file.stem
    globals()[df_name] = pd.read_csv(csv_file)
    print(f"Loaded {csv_file.name} -> {df_name}")
```
